```python
import requests, sys, time, sys, datetime, json, pandas as pd
requests.packages.urllib3.disable_warnings()
# Configuración de endpoints
url_history = "https://query1.finance.yahoo.com/v8/finance/chart/STOCK_ID.SN?symbol=STOCK_ID.SN&period1=TIME_INI&period2=TIME_END&interval=PERIOD&includePrePost=true&events=div%7Csplit%7Cearn&lang=en-US&region=US&crumb=Van3CqzKPkq&corsDomain=finance.yahoo.com"
headers = {'User-Agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 11_0 like Mac OS X) AppleWebKit/604.1.38 (KHTML, like Gecko) Version/11.0 Mobile/15A356 Safari/604.1'}

# Lista de nemotécnicos del IPSA & IGPA
stocks = [
{"code" : "ENELAM" , "name" : "ENEL AMERICAS S.A." } ,
{"code" : "CHILE" , "name" : "BANCO DE CHILE" } ,
{"code" : "SQM-B" , "name" : "SOC QUIMICA MINERA DE CHILE S.A. SERIE B" } ,
{"code" : "BSANTANDER" , "name" : "BANCO SANTANDER-CHILE" } ,
{"code" : "COPEC" , "name" : "EMPRESAS COPEC S.A." } ,
{"code" : "FALABELLA" , "name" : "FALABELLA S.A." } ,
{"code" : "CMPC" , "name" : "EMPRESAS CMPC S.A." } ,
{"code" : "ENELCHILE" , "name" : "ENEL CHILE S.A." } ,
{"code" : "CENCOSUD" , "name" : "CENCOSUD S.A." } ,
{"code" : "BCI" , "name" : "BANCO DE CREDITO E INVERSIONES" } ,
{"code" : "PARAUCO" , "name" : "PARQUE ARAUCO S.A." } ,
{"code" : "COLBUN" , "name" : "COLBUN S.A." } ,
{"code" : "CCU" , "name" : "COMPANIA CERVECERIAS UNIDAS S.A." } ,
{"code" : "AGUAS-A" , "name" : "AGUAS ANDINAS S.A., SERIE A" } ,
{"code" : "ENTEL" , "name" : "EMP. NACIONAL DE TELECOMUNICACIONES S.A." } ,
{"code" : "CENCOSHOPP" , "name" : "CENCOSUD SHOPPING S.A." } ,
{"code" : "ANDINA-B" , "name" : "EMBOTELLADORA ANDINA S.A. SERIE B" } ,
{"code" : "CONCHATORO" , "name" : "VINA CONCHA Y TORO S.A." } ,
{"code" : "ECL" , "name" : "ENGIE ENERGIA CHILE S.A." } ,
{"code" : "CAP" , "name" : "CAP S.A." } ,
{"code" : "MALLPLAZA" , "name" : "PLAZA S.A." } ,
{"code" : "ITAUCORP" , "name" : "ITAU CORPBANCA" } ,
{"code" : "IAM" , "name" : "INVERSIONES AGUAS METROPOLITANAS S.A." } ,
{"code" : "AESGENER" , "name" : "AES GENER S.A." } ,
{"code" : "SMU" , "name" : "SMU S.A." } ,
{"code" : "SONDA" , "name" : "SONDA S.A." } ,
{"code" : "RIPLEY" , "name" : "RIPLEY CORP S.A." } ,
{"code" : "VAPORES" , "name" : "COMPANIA SUD AMERICANA DE VAPORES S.A." } ,
{"code" : "ILC" , "name" : "INVERSIONES LA CONSTRUCCION S.A." } ,
{"code" : "SECURITY" , "name" : "GRUPO SECURITY S.A." },
# OTROS
{"code" : "LTM" , "name" : "LATAM AIRLINES GROUP S.A." },
{"code" : "ENJOY" , "name" : "ENJOY" }]

today_unix = int(datetime.datetime.today().timestamp())
last_360d_unix = int((datetime.datetime.today() - datetime.timedelta(days=360)).timestamp())
today = datetime.datetime.today().strftime('%Y%m%d')
for stock in stocks:    
    print("Download 1d",stock["code"])
    # Descargamos json raw
    r = requests.get(url_history.replace("STOCK_ID",stock["code"]).replace("TIME_INI",str(last_360d_unix)).replace("TIME_END",str(today_unix)).replace("PERIOD","1d"), timeout=10, headers=headers, allow_redirects=True, verify=False)
    open('data/raw_daily_'+stock["code"]+"_5m_"+today+".json", 'wb').write(r.content)
    
    # Leemos json raw
    data = json.load(open('data/raw_daily_'+stock["code"]+"_5m_"+today+".json"))
    data_structure = []
    
    # Transformamos la estructura de datos
    timestamp = data["chart"]["result"][0]["timestamp"]
    indicators = data["chart"]["result"][0]["indicators"]["quote"][0]
    for i in range(0,len(timestamp)):
        # Estructura normalizada
        record = {}
        record["date"] = datetime.datetime.fromtimestamp(timestamp[i]).strftime('%Y-%m-%d')
        record["open"] = "{:.3f}".format(indicators["open"][i])
        record["close"] = "{:.3f}".format(indicators["close"][i])
        record["volume"] = indicators["volume"][i]
        record["low"] = "{:.3f}".format(indicators["low"][i])
        record["high"] = "{:.3f}".format(indicators["high"][i])
        
        data_structure.append(record)
    
    # Estructura como DataFrame 
    df = pd.DataFrame(data_structure)
    display(df)
    df.to_excel("data/"+stock["code"]+".xlsx")
    time.sleep(0.5)
```
